```python
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm
import numpy as np
import matplotlib.pyplot as plt
import os
import matplotlib.font_manager as fm
from matplotlib.font_manager import FontProperties
font_path = '/data/workspace/yancheng/MM/Cache_Model/data/results/rbchigh_recuent/rbchigh_recuent/results/rbchigh_recuent/sv/times.ttf'
font_properties = FontProperties(fname=font_path, size=20)

colors = ['#a6cee3', '#1f78b4', '#b2df8a', '#33a02c', '#fb9a99', '#e31a1c', '#fdbf6f', '#ff7f00', '#cab2d6', '#6a3d9a']
rmse = []
l2_error = []
relative_l2_error = []
train_loss = []
vali_loss = []

with open('/data/workspace/yancheng/MM/neural_manifold_operator/plot/ns1e5.log', 'r') as file:
    lines = file.readlines()
    for line in lines:
        if 'RMSE:' in line:
            rmse.append(float(line.split('RMSE:')[1].split(',')[0].strip()))
            l2_error.append(float(line.split('L2 error:')[1].split(',')[0].strip()))
            relative_l2_error.append(float(line.split('Relative L2 error:')[1].split(',')[0].strip()))
        elif 'Epoch:' in line:
            train_loss.append(float(line.split('Train Loss:')[1].split('Vali Loss:')[0].strip()))
            vali_loss.append(float(line.split('Vali Loss:')[1].strip()))



line_width = 3
plt.figure(figsize=(10, 5))
plt.plot(rmse, label='RMSE', color=colors[0], linewidth=line_width)
plt.plot(l2_error, label='L2 Error', color=colors[1],linewidth=line_width)
plt.plot(relative_l2_error, label='Relative L2 Error', color=colors[2],linewidth=line_width)
plt.fill_between(range(len(rmse)), rmse, alpha=0.1, color=colors[0])
plt.fill_between(range(len(l2_error)), l2_error, alpha=0.1, color=colors[1])
plt.fill_between(range(len(relative_l2_error)), relative_l2_error, alpha=0.1, color=colors[2])
#plt.title('RMSE, L2 Error, and Relative L2 Error over Epochs', fontproperties=font_properties, fontsize=14)
plt.xlabel('Epoch', fontproperties=font_properties)
plt.ylabel('Value', fontproperties=font_properties)
plt.xticks(fontproperties=font_properties)
plt.yticks(fontproperties=font_properties)
plt.legend(prop=font_properties)
#plt.grid(color='lightblue', linestyle='-', linewidth=0.5) # 
plt.savefig('NS_rmse_l2_relative_l2.png', dpi=500) # 
plt.show()
plt.figure(figsize=(10, 5))
plt.plot(train_loss, label='Train Loss', color=colors[7],linewidth=line_width)
plt.plot(vali_loss, label='Validation Loss', color=colors[6],linewidth=line_width)
plt.fill_between(range(len(train_loss)), train_loss, alpha=0.1, color=colors[7])
plt.fill_between(range(len(vali_loss)), vali_loss, alpha=0.1, color=colors[6])
#plt.title('Train and Validation Loss over Epochs', fontproperties=font_properties, fontsize=14)
plt.xlabel('Epoch', fontproperties=font_properties)
plt.ylabel('Loss', fontproperties=font_properties)
plt.xticks(fontproperties=font_properties)
plt.yticks(fontproperties=font_properties)
plt.legend(prop=font_properties)
plt.savefig('NS_train_val_loss.png', dpi=500) # 
plt.show()
```
